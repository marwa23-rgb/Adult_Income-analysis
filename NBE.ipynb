```python
import pandas as pd
import os

# List of CSV files
csv_files = ['سوق راس المال.csv', 'شهادات استثمار.csv', '1شهادات ادخار بالعمله المحليه.csv','1شهادات استثمار بالعمله الاجنبيه.csv','Tحسابات التوفير.csv','papers.csv','افتح حسابك.csv','الودائع بالعمله الاجنبيه.csv','الودائع بالعمله المحلية.csv','بطاقات ائتمان.csv','حساب الأهلي اليومي2.csv',
'حساب جاري.csv','حساب فايدة بلس .csv','حسابات التوفير.csv','خدمات الحفظ.csv','شركات.csv','شهادات الاستثمار.csv','صناديق الاستثمار.csv']
# Specify the directory where your CSV files are located (if needed)
data_directory = '/content/Folder/'  # Replace with the actual directory

# Initialize an empty list
dataframes = []

# Iterate over the files
for file in csv_files:
    file_path = os.path.join(data_directory, file)
    try:
        # Try reading the CSV file with UTF-8 encoding
        df = pd.read_csv(file_path, encoding='utf-8')
        dataframes.append(df)
    except pd.errors.ParserError as e:
        print(f"Error reading {file}: {e}")
        # Handle the error, e.g., try a different encoding or skip the file

# Concatenate all dataframes
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined dataframe
combined_df.to_csv('combined_file.csv', index=False, encoding='utf-8')

```
